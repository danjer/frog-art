```python
%load_ext autoreload
%autoreload 2

from model import EmbeddingModel
from tqdm.notebook import tqdm
import os
import numpy as np
```

```python
embedder = EmbeddingModel()
```

```python
embeddings_list = []
ids_list = []

for file in tqdm(os.listdir("images")):
    embeddings_list.append(embedder.embed_image(os.path.join("images", file)))
    ids_list.append(file.split(".")[0])  # Use the filename without extension as ID
```

```python
import chromadb

client = chromadb.HttpClient(host="localhost", port=8000)
client.delete_collection("embeddings")
collection = client.get_or_create_collection("embeddings", embedding_function=None)

collection.add(
    embeddings=np.concatenate(embeddings_list),
    ids=ids_list,
)
```

```python
collection.query(query_embeddings=embeddings_list[2], n_results=3)
```
